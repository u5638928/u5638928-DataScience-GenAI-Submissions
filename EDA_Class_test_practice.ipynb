# Task
Analyze the `df` DataFrame to calculate the average 'salary_in_usd' for each 'job_title', identify the top 10 job titles with the highest average salaries, visualize these top 10 job titles and their average salaries using a bar chart, and provide insights into the findings.

## Calculate Average Salary by Job Title

### Subtask:
Group the DataFrame by 'job_title' and calculate the average 'salary_in_usd' for each job title.


**Reasoning**:
To calculate the average salary for each job title, I will group the DataFrame by 'job_title' and then compute the mean of 'salary_in_usd'.



```python
average_salary_by_job_title = df.groupby('job_title')['salary_in_usd'].mean()
print("Average salary by job title calculated.")
print(average_salary_by_job_title.head())
```

## Identify Top 10 Job Titles

### Subtask:
Sort the average salaries by job title in descending order and select the top 10 job titles with the highest average salaries.


**Reasoning**:
To identify the top 10 job titles with the highest average salaries, I will sort the `average_salary_by_job_title` Series in descending order and then select the first 10 entries.



```python
top_10_job_titles_salary = average_salary_by_job_title.sort_values(ascending=False).head(10)
print("Top 10 job titles with highest average salaries:")
print(top_10_job_titles_salary)
```

## Visualize Top 10 Job Titles by Average Salary

### Subtask:
Create a bar chart to visualize the top 10 job titles with the highest average salaries, including appropriate labels and a title.

**Reasoning**:
To visualize the top 10 job titles with the highest average salaries, I will use seaborn to create a bar plot from the `top_10_job_titles_salary` Series, ensuring clear labels and a title.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the bar plot
plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_job_titles_salary.values, y=top_10_job_titles_salary.index, palette='viridis')

# Set labels and title
plt.xlabel('Average Salary in USD')
plt.ylabel('Job Title')
plt.title('Top 10 Job Titles by Average Salary in USD')

# Adjust layout to prevent labels from overlapping
plt.tight_layout()

# Display the plot
plt.show()
```

**Reasoning**:
The previous code generated a `FutureWarning` because `palette` was used without `hue`. To address this, I will assign the `y` variable to `hue` and set `legend=False` as suggested by the warning, which will also ensure distinct colors for each bar.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the bar plot
plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_job_titles_salary.values, y=top_10_job_titles_salary.index, hue=top_10_job_titles_salary.index, palette='viridis', legend=False)

# Set labels and title
plt.xlabel('Average Salary in USD')
plt.ylabel('Job Title')
plt.title('Top 10 Job Titles by Average Salary in USD')

# Adjust layout to prevent labels from overlapping
plt.tight_layout()

# Display the plot
plt.show()
```
